```python
#!ipi install requests
import requests
import sys

# Pegue sua chave aqui: http://openweathermap.org
CHAVE = "83f59a0bdac3988b65c02eec05a4d758"
URL = "http://api.openweathermap.org/data/2.5/weather"

CIDADE = "São Paulo"
PAIS = "br"

result = requests.get(URL, params={"q": "{},{}".format(CIDADE, PAIS),
                                   "appid": CHAVE,
                                   "units": "metric",
                                   "lang": "pt_br"})

if result.status_code == 200:
    previsão = result.json()
    print("Cidade: {name} Atual: {main[temp]}°C Mínima: {main[temp_min]}°C "
          "Máxima: {main[temp_max]}°C Humidade: {main[humidity]}%".format(**previsão))
else:
    print("Erro: {}".format(result.status_code), file=sys.stderr)
```
